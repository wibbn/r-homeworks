## 1

```r
library("openxlsx")
library("dplyr")

df <- read.xlsx("./data/practice4/GAZ.xlsx")

colnames(df) <- c("Дата замера", "Давление", "Температура", "Добыча", "Конденсат", "Вода", "ID", "Куст", "Группа")
df[1] <- sapply(df[1], function(x) { ifelse(is.na(as.numeric(x)), as.character(as.Date(x, format="%d/%m/%Y")), as.character(convertToDate(x, origin = "1900-01-01"))) })

df$ID <- factor(df$ID)
df$Куст <- factor(df$Куст)
df$Группа <- factor(df$Группа)

df <- transform(df, Давление = as.numeric(Давление))
df <- transform(df, Температура = as.numeric(Температура))
df <- transform(df, Добыча = as.numeric(Добыча))
df <- transform(df, Конденсат = as.numeric(Конденсат))
df <- transform(df, Вода = as.numeric(Вода))

head(df)
tail(df)
```

## 2

```r
df <- df[rowSums(is.na(df)) == 0,]
head(df)
```

## 3

```r
df <- transform(df, Температура = Температура + 273.15)
head(df)
```

## 5

```r
df["НефтиНаКонденсат"] <- df["Добыча"] / df["Конденсат"]
df["НефтиНаВоду"] <- df["Добыча"] / df["Вода"]
df["ВодыНаКонденсат"] <- df["Вода"] / df["Конденсат"]

head(df)
```

## 6

```r
getYear <- function(dateString) {
    return (as.numeric(format(as.Date(dateString),'%Y')))
}

df_2018 <- filter(df, getYear(Дата.замера) == 2018)
mining <- aggregate(Добыча ~ Куст, df_2018, sum)

head(df_2018)
mining
```
